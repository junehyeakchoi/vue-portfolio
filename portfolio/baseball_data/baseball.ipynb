```python
def getMeanSafety(seaon, name, category, flag):
    if flag == 0:
        temp = df[(df["선수명"]==name) & (df["시즌"]==season)]
    elif flag == 1:
        temp = df[(df["선수명"]==name) & (df["시즌"]>=season)]
    elif flag == 2:
        temp = df[(df["선수명"]==name) & (df["시즌"]<=season)]
    else: 
        return
    return temp[category].mean()
```

```python
def testtemp(seson,name,category,flag):
    if flag == 0:
        temp = df[(df["팀명"]==name) & (df["시즌"]==season)]
    elif flag == 1:
        temp = df[(df["팀명"]==name) & (df["시즌"]>=season)]
    elif flag == 2:
        temp = df[(df["팀명"]==name) & (df["시즌"]<=season)]
    else: 
        return
    return temp[category].mean()
```
